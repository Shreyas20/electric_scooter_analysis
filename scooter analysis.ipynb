```python
window=113 #max time a trip lasted
start=A_df['StartTime'].min()
end=A_df['StartTime'].max()
current=start
windowL=[]
while(current<end):
    windowL.append(((A_df[A_df['StartTime'].between(current,current + datetime.timedelta(minutes=window))].agg({'TripID':'count', 'ScooterID':'nunique'}))['TripID'],
     (A_df[A_df['StartTime'].between(current,current + datetime.timedelta(minutes=window))].agg({'TripID':'count', 'ScooterID':'nunique'}))['ScooterID']))
    current=current + datetime.timedelta(minutes=window)
    
```

```python
res = max(windowL, key = itemgetter(1))
res
```

So only 111 scooters are operating when there maximum density of the operational scooters. That means, at any time only 16% of the maximum no of scooters are operational in the best case.

    no of estimated scooter ~ 240 <br>
